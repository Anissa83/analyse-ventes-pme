```python
import pandas as pd

#Analyse des données avec pandas

fichier_ventes = "Donnés brief data engineer - ventes.csv"  
fichier_produits = "Donnés brief data engineer - produits.csv"  
fichier_magasins = "Donnés brief data engineer - magasins.csv"  

# Charger les fichiers CSV avec pandas
df_ventes = pd.read_csv(fichier_ventes)
df_produits = pd.read_csv(fichier_produits)
df_magasins = pd.read_csv(fichier_magasins)
```

```python
#Fusionner les données
df_ventes_produits = df_ventes.merge(df_produits, on="ID Référence produit", how="left")

df_ventes_produits
```

```python
#Analyse des ventes
df_ventes_produits["chiffre_affaires"] = df_ventes_produits["Quantité"] * df_ventes_produits["Prix"]
chiffre_affaires_total = df_ventes_produits["chiffre_affaires"].sum()
print(f"Chiffre d'affaires total : {chiffre_affaires_total} €")
```

```python
# Produits les plus vendus
top_produits = df_ventes_produits.groupby("Nom")["Quantité"].sum().sort_values(ascending=False)
print(top_produits.head(5))
```

```python
#Visualisation des résultats
import matplotlib.pyplot as plt
import seaborn as sns

# Top 5 produits les plus vendus
plt.figure(figsize=(10,5))
sns.barplot(x=top_produits.index[:5], y=top_produits.values[:5])
plt.title("Top 5 des produits les plus vendus")
plt.xlabel("Produit")
plt.ylabel("Quantité vendue")
plt.xticks(rotation=45)
plt.show()

```
